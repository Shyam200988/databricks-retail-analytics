```python
base_path = "/Volumes/workspace/bronze/retail_analytics_volume/"
```

```python
# Read the customers.csv file from the Unity Catalog volume
# Tell Spark the file format is CSV
customers_df = (
    spark.read.format("csv")
    .option("header", "true")  # First row contains column names
    .option("inferSchema", "true")  # Spark automatically detects column data types
    .load(base_path + "customers.csv")
)  

```

```python
# Write the dataframe to a Delta table inside the bronze schema

customers_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze.customers")
```

```python
# Read products.csv from the volume

products_df = spark.read.format("csv") \
.option("header","true") \
.option("inferSchema","true") \
.load(base_path + "products.csv")

```

```python
# Write products dataframe into bronze schema as Delta table

products_df.write.format("delta") \
.mode("overwrite") \
.saveAsTable("bronze.products")
```

```python
# Read orders dataset from volume

orders_df = spark.read.format("csv") \
.option("header","true") \
.option("inferSchema","true") \
.load(base_path + "orders.csv")

```

```python
# Write orders dataframe into bronze schema as Delta table

orders_df.write.format("delta").option("mergeSchema", "true") \
.mode("overwrite") \
.saveAsTable("bronze.orders")
```
