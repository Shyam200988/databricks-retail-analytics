```python
# Load the main fact table created in the Silver layer

sales_df = spark.table("silver.sales_fact")

display(sales_df)
```

```python
from pyspark.sql.functions import sum

product_sales = sales_df.groupBy(
    "product_id",
    "product_name",
    "product_category"
).agg(
    sum("quantity").alias("total_quantity_sold"),
    sum("revenue").alias("total_revenue")
)

display(product_sales)
```

```python
product_sales.write.format("delta") \
.mode("overwrite") \
.saveAsTable("gold.product_sales")
```

```python
from pyspark.sql.functions import concat_ws,col

customer_sales = sales_df.groupBy(
    "customer_id",
    (concat_ws(" ", col("first_name"), col("last_name"))).alias("customer_name"),
    "customer_city"
).agg(
    sum("revenue").alias("total_customer_revenue")
)

display(customer_sales)
```

```python
customer_sales.write.format("delta") \
.mode("overwrite") \
.saveAsTable("gold.customer_sales")
```

```python
daily_sales = sales_df.groupBy(
    "order_date"
).agg(
    sum("revenue").alias("daily_revenue")
)

display(daily_sales)
```

```python
daily_sales.write.format("delta") \
.mode("overwrite") \
.saveAsTable("gold.daily_sales")
```

```python
city_sales = sales_df.groupBy(
    "store_city"
).agg(
    sum("revenue").alias("total_city_revenue")
)

display(city_sales)
```

```python
city_sales.write.format("delta") \
.mode("overwrite") \
.saveAsTable("gold.city_sales")
```
